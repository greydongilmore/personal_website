# Support Vector Machine

A support vector machine (SVM) is an extension of LR in that a line is used to model the data. However, in the SVM two additional lines are added to either side of the middle line. These other lines are termed hyperplanes and they provide a decision boundary to separate datapoints. If the number of features in the dataset is 2 then the hyperplane is a linear line. However, if the number of features is 3 or more than the hyperplane becomes a complex n-dimensional plane. The hyperplanes are formed by using support vectors, which are data points closest to the line. 

With SVM regression (SVR) an accuracy term is applied to the line called epsilon. The epsilon is the error threshold used when attempting to fit the hyperplane to the datapoints. Datapoints with an error below the threshold do not contribute to the regression fit, while datapoints with error beyond the threshold contribute linearly.
