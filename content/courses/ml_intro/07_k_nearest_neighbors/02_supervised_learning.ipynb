# K-nearest neighbors

The K-Nearest Neighbor (KNN) algorithm uses feature similarity to assign new data points to existing clusters of data. The similarity of a new data point to existing clusters of data is most often measured using a distance metric such as Euclidean distance.

```python
from sklearn.neighbors import KNeighborsRegressor

# Determine the optimal number of neighbours to use
error_rate = []
for i in range(1, 21):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.sqrt(mean_squared_error(y_test.ravel(),pred_i)))
    print('Done neighbor {}'.format(i))

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(list(range(1, 21)), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
ax.set_xticks(list(range(1, 21)))
ax.set_xticklabels(list(range(1, 21)))
plt.title('RMSE vs. K Neighbors', fontweight='bold')
plt.xlabel('K Neighbors', fontweight='bold')
plt.ylabel('RMSE value', fontweight='bold')
plt.show()
```
